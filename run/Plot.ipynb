```python
# Import the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
ux = pd.read_csv('/Users/nirupam/Desktop/Programming/Lattice Boltzmann Method/High Fidelity Lattice Boltzmann Method/2D_Lid_Driven_Cavity/run/ux.csv', header=None)
uy = pd.read_csv('/Users/nirupam/Desktop/Programming/Lattice Boltzmann Method/High Fidelity Lattice Boltzmann Method/2D_Lid_Driven_Cavity/run/uy.csv', header=None)
```

```python
ux = ux.to_numpy()
uy = uy.to_numpy()
```

```python
x = np.linspace(0, 1, len(ux))
y = np.linspace(0, 1, len(uy))
X, Y = np.meshgrid(x, y)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
plt.contourf(X, Y, ux, cmap='rainbow')
plt.xlabel(r'$x$', fontsize=15)
plt.ylabel(r'$y$', fontsize=15)
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
plt.contourf(X, Y, uy, cmap='rainbow')
plt.xlabel(r'$x$', fontsize=15)
plt.ylabel(r'$y$', fontsize=15)
plt.colorbar()
plt.show()
```

```python
u_ghia = np.loadtxt('/Users/nirupam/Desktop/Programming/Lattice Boltzmann Method/High Fidelity Lattice Boltzmann Method/2D_Lid_Driven_Cavity/run/ghia_data_u_vs_y.txt').T
v_ghia = np.loadtxt('/Users/nirupam/Desktop/Programming/Lattice Boltzmann Method/High Fidelity Lattice Boltzmann Method/2D_Lid_Driven_Cavity/run/ghia_data_v_vs_x.txt').T
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(u_ghia[0], u_ghia[1], '.', ms=15, label='Ghia et. al.')
ax.plot(y, ux[:, int(len(ux)/2)], '-', label='LBM')
ax.set_title(r'Plot for $u$ vs $y$ at $x = 0.5$', fontsize=20)
ax.set_xlabel(r'$y$', fontsize=20)
ax.set_ylabel(r'$u$', fontsize=20)
plt.legend(fontsize=15)
ax.tick_params(labelsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(v_ghia[0], v_ghia[1], '.', ms=15, label='Ghia et. al.')
ax.plot(x, uy[int(len(ux)/2), :], '-', label='LBM')
ax.set_title(r'Plot for $v$ vs $x$ at $y = 0.5$', fontsize=20)
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'$v$', fontsize=20)
plt.legend(fontsize=15)
ax.tick_params(labelsize=15)
plt.show()
```
